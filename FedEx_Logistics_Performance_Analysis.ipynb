# Project Summary

This project focuses on analyzing FedEx Logistics shipment data to gain insights into supply chain performance, delivery efficiency, and transportation costs. The dataset consists of 10,324 records and 33 variables containing information related to purchase orders, shipment modes, vendor agreements, delivery schedules, product details, freight costs, and destination countries. The primary objective of this analysis is to identify factors affecting delivery performance, understand cost drivers, and provide actionable recommendations to improve logistics operations.

The project began with data cleaning and preprocessing. Missing values were examined carefully, and appropriate techniques were applied based on the nature of each variable. For example, missing values in the Line Item Insurance column were replaced with zero where insurance was not applicable, while categorical values such as "Unknown" in the Dosage column were retained to preserve information. Date columns were converted into datetime format to enable time-based analysis and feature engineering.

Several new variables were created to enhance the analysis. Delivery Delay Days was calculated by comparing the scheduled delivery date with the actual delivery date. Vendor Lead Time and Order Cycle Time were also derived to measure the efficiency of different stages of the supply chain. These engineered features helped in evaluating delivery performance and identifying operational bottlenecks.

Exploratory Data Analysis (EDA) was conducted using Pandas, NumPy, Matplotlib, Seaborn, and Plotly. Various visualizations were created to understand shipment distribution across countries, transportation modes, vendors, and manufacturing sites. Relationships between freight cost, shipment weight, quantity, and delivery timelines were also examined. Outlier detection techniques were used to identify unusually expensive or heavy shipments that could impact logistics costs.

The analysis revealed several important insights. Air transportation was found to be the most frequently used shipment mode, highlighting its significance in global logistics operations. Freight costs generally increased with shipment weight, indicating a strong relationship between transportation expenses and shipment size. Certain countries accounted for a large proportion of total shipments, helping identify key operational regions. Delivery performance varied across shipment modes and vendor agreements, suggesting opportunities for process optimization and better resource allocation.

Based on these findings, recommendations were proposed to improve supply chain efficiency. These include monitoring high-cost shipment routes, optimizing transportation methods, improving vendor coordination, and focusing on regions with frequent delivery delays. The insights generated from this project can support decision-making, reduce operational costs, enhance delivery performance, and improve overall customer satisfaction.

Overall, this project demonstrates how data analytics can be applied to logistics and supply chain management to uncover meaningful patterns, improve operational efficiency, and support strategic business decisions through data-driven insights.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df.head()
````

````python
# Filling missing value for shipment mode - use mode because it's a categorical column
df['Shipment Mode'] = df['Shipment Mode'].fillna(df['Shipment Mode'].mode()[0])
````

````python
df['Shipment Mode'].head()
````

````python
# Filling missing dosage as Unknown
df['Dosage'] = df['Dosage'].fillna("Unknown")
df['Dosage']
````

````python
# verify by checking whether shipments with missing insurance still have freight costs and values
df[['Line Item Value',
    'Freight Cost (USD)',
    'Line Item Insurance (USD)']].head()
````

````python
df['Line Item Insurance (USD)'] = \
df['Line Item Insurance (USD)'].fillna(0)

# Insurance cost cannot exist if insurance was not taken. In many logistics datasets,
# missing insurance effectively means zero insurance cost
````

````python
df["Line Item Insurance (USD)"].head(25)
````

````python
df.isnull().sum()
````

````python
# Measures whether the shipment arrived before or after the planned date.
df['Delivery_Delay_Days'] = (
    pd.to_datetime(df['Delivered to Client Date']) -
    pd.to_datetime(df['Scheduled Delivery Date'])
).dt.days
````

````python
# Time taken from sending PO to the vendor until delivery.
df['Vendor_Lead_Time'] = (
    pd.to_datetime(df['Delivered to Client Date'].replace(['Date Not Captured', 'N/A - From RDC'], np.nan)) -
    pd.to_datetime(df['PO Sent to Vendor Date'].replace(['Date Not Captured', 'N/A - From RDC'], np.nan))
).dt.days
````

````python
# On-Time Delivery Flag
df['On_Time_Delivery'] = np.where(
    df['Delivery_Delay_Days'] <= 0,
    'Yes',
    'No'
)
````

### What all manipulations have you done and insights you found?

* Handled missing values by analyzing each column separately. For example, missing values in Line Item Insurance (USD) were replaced with 0, while categorical fields such as Dosage were kept as "Unknown" to preserve information.
* Converted date columns into datetime format and created new features such as Delivery Delay Days, Vendor Lead Time, and On-Time Delivery Status for better business analysis.
* Checked for duplicates and data inconsistencies to ensure data quality before performing analysis.
